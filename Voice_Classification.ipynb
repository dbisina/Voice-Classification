```python
#extract the targety cvariable with an x
X = df.iloc[:, :-1]
print(df.shape)
print(X.shape)
```

```python
#converting the labels to numerical form
from sklearn.preprocessing import LabelEncoder
```

```python
y = df.iloc[:, -1]

gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
y
```

```python
#to make the model efficient we have to scale the model between 1 and o
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
SVC_model = SVC()
SVC_model.fit(X_train, y_train)
```

```python
y_pred = SVC_model.predict(X_test)
```

```python
print('Accuracy Score: ')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
#Tuning the parameters to get even better
#Optimizi=g the
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C' : [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001]}
```

```python
grid = GridSearchCV(SVC(), param_grid,refit=True,verbose=2)
grid.fit(X_train, y_train)
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print('Accuracy score')
print(metrics.accuracy_score(y_test,grid_predictions))
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
